```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Iris.csv")
df = pd.DataFrame(data)
(df.head())
```

```python
(df.describe())#details about data
```

```python
df.info()#information about data types and null values
```

```python
x=df["Species"].values
y=df["SepalWidthCm"].values
plt.bar(x, y)
plt.title("Sepal width of different species")
plt.xlabel("Sepal length(cm)")
plt.ylabel("Sepal width(cm)")
```

```python
x=df["Species"].values
y=df["SepalLengthCm"].values
plt.bar(x, y)
plt.title("Sepal width of different species")
plt.xlabel("Sepal length(cm)")
plt.ylabel("Sepal width(cm)")
```

```python
df["Species"].value_counts()
```

```python
species_count = df["Species"].value_counts()
plt.pie(species_count,labels=species_count.index,autopct='%1.1f%%',startangle=90)
```
