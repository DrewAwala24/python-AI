```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("transnzoiarainfall.csv")
```

```python
df = pd.DataFrame(data)
#Cleaning data
print(df.isnull().sum())#checks for missing values in the dataset
df.fillna(0, inplace=True)#replaces missing values with 0
#df.dropna(inplace=True) #removes any rows with missing values
print(df)

```

```python
print(df.head())
```

```python
average = np.mean(df["Rainfall_mm"])
print("Average Rainfall: ",average)
```

```python
x = df["Month"]
y = df["Rainfall_mm"]
plt.bar(x,y)
plt.title("Rainfall in Trans Nzoia")
plt.xlabel("Months")
plt.ylabel("Rainfall_mm")
plt.show
```
